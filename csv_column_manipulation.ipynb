```
import numpy as np
import pandas as pd
```

```
data = pd.read_excel('finaljazz.xlsx')
```

```
print(data)
```

```
print(np.shape(data))
```

```
%%time
length = np.shape(data)[0]
uniquelist = []
yearlist = []
for i in range(0,len):
  uniquelist.append(i)
  uniquelist[i] = []
 

print(length)

length = 500
for i in range(0,length):
  ownrid = data['rid'][i]
  ownmusicid = data['musicid'][i]
  ownyear = data['year'][i]
  for j in range(i+1,len):
    otherrid = data['rid'][i]
    othermusicid = data['musicid'][i]
    otheryear = data['year'][i]
    
    flag =0
    for ele in ownmusicid.split('|'):
      if ele in othermusicid.split('|') and ele != '0' and ele != 0:
        flag =1
    
    if flag ==1 and ownrid != otherrid and ownyear == otheryear: #len(ownrid) != 0 and len(otherrid) != 0 and  #no need to do seperately as flag will handle
      uniquelist[i].append(data['location'][j])
      yearlist.append(ownyear)
      
  if i % 10==0:
    print(i)
    
    
# for i in range(0,length):    
#   print(uniquelist[i])
```

```
finallist1 = []
finallist2 = []

for i,templist in enumerate(uniquelist):
  templen =  np.shape(templist)[0]
  ownname = data['location'].iloc[i]
  for j in range(0,templen):
    othername = uniquelist[i][j]
    finallist1.append(ownname)
    finallist2.append(othername)   
```

```
print(np.shape(finallist1))
print(np.shape(finallist2))

# for i in range(0,np.shape(finallist1)[0]):
#   print(finallist1[i],finallist2[i])
  
print(finallist1)
print(finallist2)
```

```
newlist = []
for i in range(0,np.shape(finallist1)[0]):
  newlist.append(i)
  newlist[i] = []

for i in range(0,np.shape(finallist1)[0]):
  newlist[i].append(finallist1[i])
  newlist[i].append(finallist2[i])
  newlist[i].append(yearlist[i])

print(newlist)
```
